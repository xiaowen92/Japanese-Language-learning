### 📘 1. 过去式的区别：`かった` vs. `でした`

**1. 核心逻辑 (The Core Logic)**
日语中表达“过去是……”时，不能只凭感觉混用。决定因素在于**单词的词性**。

* **A 类选手 (い形容词):** 自带变身能力，词尾发生变化。
* **B 类选手 (名词 & な形容词):** 需要借助外部的词 (`desu` -> `deshita`) 来表示过去。

**2. 语法规则 (Syntax & Rules)**

**Mode A: い形容词 (i-Adjectives)**

* **规则:** 去掉词尾的 `い`，加上 `かった` (katta)。
* *公式:* `[词干] + かった`

**Mode B: 名词 (Nouns) / な形容词 (na-Adjectives)**

* **规则:** 礼貌体中，直接把结尾的 `です` (desu) 换成 `でした` (deshita)。
* *公式:* `[词] + でした`

**3. 典型示例 (Examples)**

**Example A: い形容词 (自带变身)**

* `暑い` (热) → `暑**かった**` (过去很热)
* `楽しい` (开心) → `楽し**かった**` (过去很开心)
* *注意:* 不要说 `暑いでした` (❌ 错误)。

**Example B: 名词 / な形容词 (借助 `でした`)**

* `雨` (雨 - 名词) → `雨**でした**` (过去是雨/下过雨)
* `元気` (健康 - な形容词) → `元気**でした**` (过去很健康/很好)
